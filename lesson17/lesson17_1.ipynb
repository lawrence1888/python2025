```python
#在終端機上打 pip install -r requirements.txt

import pandas as pd

df = pd.read_csv('各鄉鎮市區人口密度.csv')
df
```

```python
df.info()
```

```python
df1 = df.rename(columns={'statistic_yyy':'統計年',
                  'site_id':'區域別',
                  'people_total':'年底人口數',
                  'area':'土地面積',
                  'population_density':'人口密度'
                    })
df1
```

```python
#axis=0 表示操作的軸是「列」，也就是針對資料的每一行進行判斷和刪除。
#how='any' 表示只要該列中有任意一個缺失值（NaN），整列就會被刪除。


df2 = df1.drop(index=0)
df3=df2.dropna(axis=0,how='any')
```

```python
df4 = df3.drop('統計年',axis=1)
df4
```

```python
df4.info()
```

```python
for value in df4['年底人口數']:
    print(value)
```

```python
#轉換資料類型
#strToInt：將字串轉成整數，失敗就回傳 0。
#apply(strToInt)：對兩個欄位（年底人口數、人口密度）做轉換。
#astype('float')：將土地面積轉成 float。

def strToInt(x:str)->int:
    try:
        return int(x)
    except:
        return 0

df4['年底人口數'] = df4['年底人口數'].apply(strToInt)
df4['人口密度'] = df4['人口密度'].apply(strToInt)
df4['土地面積']= df4['土地面積'].astype('float')
```

```python
df4.info()
```

```python
#df:定義函式
#split_area是自己創立的變項


def split_area(v:str)->str:
    if len(v) >= 6:
        return v[:3]
    else:
        return v
    
def split_area1(v:str)->str:
    if len(v) >=6:
        return v[-3:]
    else:
        return v
    

df4['縣市'] = df4['區域別'].apply(split_area)
df4['區域'] = df4['區域別'].apply(split_area1)
df4
```

```python
df5 = df4.reindex(columns=['縣市','區域','年底人口數','土地面積','人口密度'])
df5
```

```python
#搜尋
df5.query('年底人口數>=500000')
```

```python
m = df5['人口密度'].max()
df5.query('人口密度== @m')
```

```python
df5
```

```python
#agg 是 aggregate（聚合） 的縮寫，用來對分組（groupby）後的資料執行多種彙總計算。

df6 = df5.groupby(by='縣市')[['年底人口數','土地面積','人口密度']].agg([('平均','mean'),('加總','sum')])
```

```python
df6.columns.names= ["資料","統計"]
df6
```

```python
#類似excel樞紐工具的寫法(感覺像把上欄和左欄位置交換)

#stack：將 DataFrame 的指定欄位層級（columns）「往下堆疊」到行索引（index）中，將欄位轉成行索引的多層索引層級。
#unstack：將 DataFrame 的指定行索引層級「往上展開」成欄位層級，與 stack 相反。
#兩者是互為逆操作，常用於「寬資料（wide）」與「長資料（long）」格式之間轉換。

df6.stack(level=['資料','統計'],future_stack=True).unstack(level=['縣市','資料'])
```
